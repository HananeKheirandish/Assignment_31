```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker= 'o')
plt.show()
```

```python
plt.plot(ypoints, marker= '*')
plt.show()
```

```python
plt.plot(ypoints, marker = 'P')
plt.show()
```

```python
plt.plot(ypoints, marker = 'H')
plt.show()
```

```python
plt.plot(ypoints, marker= '3')
plt.show()
```

```python
plt.plot(ypoints, 'o:r')
plt.show()
```

```python
plt.plot(ypoints, 'o-.m')
plt.show()
```

```python
plt.plot(ypoints, marker= 'o', ms = 15)
plt.show()
```

```python
plt.plot(ypoints, marker= 'o', ms = 15, mec = 'r')
plt.show()
```

```python
plt.plot(ypoints, marker= 'o', ms = 15, mfc = 'r')
plt.show()
```

```python
plt.plot(ypoints, marker= 'o', ms = 15, mec = 'r', mfc = 'r')
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = '#4CAF50', mfc = '#4CAF50')
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'hotpink', mfc = 'hotpink')
plt.show()
```
